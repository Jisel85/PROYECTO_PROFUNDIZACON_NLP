```python
# TUTELA:         https://www.corteconstitucional.gov.co/relatoria/1992/T-612-92.htm
# AUTOS:          https://www.corteconstitucional.gov.co/relatoria/autos/1992/A024-92.htm
# CONSTITUCIONAL: https://www.corteconstitucional.gov.co/relatoria/1992/C-587-92.htm
def get_url_sentence(sentence):
    url_base = "https://www.corteconstitucional.gov.co/relatoria/"
    year = sentence.split("-")[-1]
    year = re.sub(r'[^a-zA-Z0-9]', "", year)  ## Have a character creep: '04Â\xa0'
    if float(year) > 91:
        year = "19" + year
    elif float(year) < 24:
        year = "20" + year    
    if sentence.lower().startswith("a"):
        url_sentence = f'{url_base}autos/{year}/{sentence}.htm'
        return url_sentence
    else:
        url_sentence = f'{url_base}{year}/{sentence}.htm'
        return url_sentence
```

```python
## Create a JSON file ##
dict_to_json = {}
with open('data/sentences.json', 'w') as file:
     json.dump(dict_to_json, file)
##
def get_json_sentences_urls(folder):
    for file in folder.iterdir():
        temp = pd.read_excel(file, skiprows=6)
        temp = temp[['sentenciav2']].dropna()
        print(f'File: {str(file).split("/")[-1]}\nwith dropna by "sentenciav2" has shape: {temp.shape}')
        temp['url_sentence'] = temp['sentenciav2'].apply(get_url_sentence)
        print(f'File: {str(file).split("/")[-1]}\nwith dropna by "sentenciav2" has shape: {temp.shape}')
        print()
        dict_sentences = {row["sentenciav2"]: row["url_sentence"] for _, row in temp.iterrows()}
        with open('data/sentences.json', 'r') as file:
              json_string = file.read()
              json_sentences = json.loads(json_string)
              json_sentences.update(dict_sentences)
        with open("data/sentences.json", "w") as file:
             file.write(json.dumps(json_sentences))
    return None

get_json_sentences_urls(MAIN_PATH / "data" / "data_sentences")
```
