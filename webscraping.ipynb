```python
import os
import json
import pandas as pd
import numpy as np
import warnings
import concurrent.futures
from pathlib import Path
##
from selenium import webdriver
from selenium.webdriver.firefox.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.firefox.firefox_profile import FirefoxProfile
from webdriver_manager.firefox import GeckoDriverManager
##
warnings.filterwarnings("ignore")
```

```python
#service = Service(executable_path=GeckoDriverManager().install())
#driver_path = GeckoDriverManager().install()
#print(driver_path)
MAIN_PATH = Path(os.getcwd())
MAIN_PATH
path_download_folder = MAIN_PATH / "data" / "downloads_files_sentences"
profile = FirefoxProfile()
options = Options()
firefox_options = webdriver.FirefoxOptions()
##
profile.set_preference("browser.download.folderList", 2)                    # 2: Downlodas in folder customizer
profile.set_preference("browser.download.dir", str(path_download_folder))   # Path folder downloads
options.profile = profile
##
firefox_options.binary_location = '/usr/bin/firefox'        ## Path firefox
firefox_options.add_argument('--headless')                  ## Don't deploy Browser
##
def open_browser():
    driver = webdriver.Firefox(
        executable_path='/home/andres-campos/.wdm/drivers/geckodriver/linux64/v0.33.0/geckodriver',
        options=firefox_options,
        firefox_profile=profile,
        )
    return driver
```

```python
with open('data/sentences.json', 'r') as file:
        json_string = file.read()
json_sentences = json.loads(json_string)
problems_urls = []

def get_sentence_by_url(sentence):
    file_path = MAIN_PATH / "data" / "downloads_files_sentences" / f'{sentence}.rtf'
    if os.path.exists(file_path):
        print(f'File: {sentence}.rtf already exist!')
        return None
    else:
        url = json_sentences[sentence]
        print(f'Working in url: {url}')
        driver = open_browser()        
        driver.get(url)
        try:
            radicado_bt = driver.find_element(By.XPATH, "/html/body/div[5]/div/div[3]/a/img")
            radicado_bt.click()
            print(f'Successfull in url: {url}')
            print()
        except:
            print(f'PROBLEMS WITH URL: {url}')
            problems_urls.append(url)
            print()
        driver.close()
        return None

#get_sentence_by_url("T-590-93")
```

```python
urls = list(json_sentences.keys())
max_threads = 8
with concurrent.futures.ThreadPoolExecutor(max_workers=max_threads) as executor:
    executor.map(get_sentence_by_url, urls)
```
