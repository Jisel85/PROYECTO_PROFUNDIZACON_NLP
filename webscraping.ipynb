```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pandas as pd
import time
```

```python
sentencias = pd.read_csv('Sentencias_Corte_1992-2022.csv',encoding='latin-1', sep=',')
```

```python
sentencias
```

```python
sentencias["Sentencia"] = "https://www.corteconstitucional.gov.co/relatoria/" + sentencias["Año"].map(str) + "/" + sentencias["Sentencia"] + ".htm" 
```

```python
sentencias
```

```python
sentencias['Sentencia'][0]
```

```python
driver = webdriver.Firefox()
for i in range(len(sentencias['Sentencia'])):
    Sentencia = sentencias['Sentencia'][i]
    driver.get(Sentencia)
    try: 
        radicado_bt = driver.find_element(By.XPATH, "/html/body/div[5]/div/div[3]/a/img")               
        radicado_bt.click()
    except:
        radicado_bt = driver.find_element(By.XPATH, "/html/body/div/div[1]/h1")
driver.close()
```
